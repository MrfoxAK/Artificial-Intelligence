```python
def cnvt_range(x):
     temp = x.split('-')
     if len(temp) == 2:
          return ((float(temp[0])) + (float(temp[1])))/2
     try:
          return float(x)
     except:
          return None
```

```python
df2['total_sqft'] = df2['total_sqft'].apply(cnvt_range)
```

```python
df2['total_sqft'] = df2['total_sqft'].dropna()
```

```python
df2['price_per_sqft'] = df2['price']*100000 / df2['total_sqft']
```

```python
df2['location'] = df2['location'].apply(lambda x:x.strip())
loc_count = df2['location'].value_counts()
```

```python
loc_count_less_10 = loc_count[loc_count<10]
```

```python
df2['location'] = df2['location'].apply(lambda x : 'other' if x in loc_count_less_10 else x)
```
