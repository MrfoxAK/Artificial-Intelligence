```python
import pandas as pd
from pandas_profiling import ProfileReport
from pydantic_settings import BaseSettings

# Step 1: Define a Settings class using BaseSettings for configuration
class Settings(BaseSettings):
    dataset_path: str = "train (1).csv"  # Path to your dataset
    report_title: str = "Pandas Profiling Report"
    save_report: bool = True  # Whether to save the report or just display it
    output_file: str = "profiling_report.html"  # The file name for the saved report

# Load the settings
settings = Settings()

# Step 2: Load your dataset using pandas
df = pd.read_csv(settings.dataset_path)

# Step 3: Generate the pandas profiling report
profile = ProfileReport(df, title=settings.report_title, explorative=True)

# Step 4: Display or save the report based on settings
if settings.save_report:
    profile.to_file(settings.output_file)
    print(f"Report saved to {settings.output_file}")
else:
    profile.to_notebook_iframe()
```

```python
from flask import Flask, render_template, request, redirect, url_for
import pandas as pd
import plotly
import plotly.express as px
import json
import os
from werkzeug.utils import secure_filename

app = Flask(__name__)
app.config['UPLOAD_FOLDER'] = 'uploads'
app.config['MAX_CONTENT_LENGTH'] = 16 * 1024 * 1024  # Max 16MB upload

# Ensure the upload folder exists
if not os.path.exists(app.config['UPLOAD_FOLDER']):
    os.makedirs(app.config['UPLOAD_FOLDER'])

ALLOWED_EXTENSIONS = {'csv'}

def allowed_file(filename):
    return '.' in filename and \
           filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        # Check if the post request has the file part
        if 'file' not in request.files:
            return render_template('index.html', error="No file part")
        file = request.files['file']
        # If user does not select file, browser may submit an empty part
        if file.filename == '':
            return render_template('index.html', error="No selected file")
        if file and allowed_file(file.filename):
            filename = secure_filename(file.filename)
            filepath = os.path.join(app.config['UPLOAD_FOLDER'], filename)
            file.save(filepath)
            # Read and analyze the CSV
            try:
                df = pd.read_csv(filepath)
                # Basic analysis
                summary = df.describe().to_html(classes='table table-striped', border=0)
                # Generate plots (example: histogram for numerical columns)
                numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
                plots = []
                for col in numeric_columns:
                    fig = px.histogram(df, x=col, title=f'Histogram of {col}')
                    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
                    plots.append({'title': f'Histogram of {col}', 'graphJSON': graphJSON})
                # For categorical columns, generate bar charts
                categorical_columns = df.select_dtypes(include=['object', 'category']).columns
                for col in categorical_columns:
                    fig = px.bar(df[col].value_counts().reset_index(),
                                 x='index',
                                 y=col,
                                 labels={'index': col, col: 'Count'},
                                 title=f'Bar Chart of {col}')
                    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
                    plots.append({'title': f'Bar Chart of {col}', 'graphJSON': graphJSON})
                # Optionally, show first few rows
                head = df.head().to_html(classes='table table-bordered', border=0)
                return render_template('results.html', tables={'Summary Statistics': summary, 'Data Head': head}, plots=plots)
            except Exception as e:
                return render_template('index.html', error=f"Error processing file: {e}")
        else:
            return render_template('index.html', error="Invalid file type. Please upload a CSV file.")
    return render_template('index.html')

if __name__ == '__main__':
    app.run(debug=True)
```
